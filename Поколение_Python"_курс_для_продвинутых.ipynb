```python
# put your python code here
def input_matrix(n, m):
    matrix = []
    for row in range(n):
        matrix.append([int(num) for num in input().split()])
    return matrix

def find_indexes_of_max(matrix, n, m):
    row_of_max = 0
    cell_of_max = 0

    for row in range(n):
        for cell in range(m):
            if matrix[row][cell] > matrix[row_of_max][cell_of_max]:
                row_of_max = row
                cell_of_max = cell
    return row_of_max, cell_of_max

n = int(input())
m = int(input())

matrix = input_matrix(n, m)
row_of_max, cell_of_x = find_indexes_of_max(matrix, n, m)
print(row_of_max, cell_of_x)
```

```python
def input_matrix(n, m):
    matrix = [[int(num) for num in input().split()] for _ in range(n)]

    return matrix


def swap_cells(matrix, cells_to_swap):
    for row in matrix:
        row[cells_to_swap[0]], row[cells_to_swap[1]] =  row[cells_to_swap[1]], row[cells_to_swap[0]]

    return matrix

def print_matrix(matrix, n, m):
    for row in matrix:
        print(*row)


n = int(input())
m = int(input())
matrix = input_matrix(n, m)
cells_to_swap = [int(cell) for cell in input().split()]
matrix = swap_cells(matrix, cells_to_swap)
print_matrix(matrix, n, m)
```

### Магический квадрат

---
Магическим квадратом порядка n называется квадратная таблица размера n×n,составленная из всех чисел 1,2,3,…,n^2 (то есть все числа разные) так, что cуммы по каждому столбцу, каждой строке и каждой из двух диагоналей равны между собой. Напишите программу, которая проверяет, является ли заданная квадратная матрица магическим квадратом.

---

```python
def input_matrix(n):
    matrix = [[int(num) for num in input().split()] for _ in range(n)]

    return matrix

def print_matrix(matrix):
    for row in matrix:
        print(*row)

def check_magic(matrix, n):
    nums_list = [i for i in range(1, n**2 + 1)]
    row_sum = sum([matrix[i][i] for i in range(n)])
    for i in range(n):
        if sum(matrix[i]) != row_sum or sum(matrix[j][i] for j in range(n)) != row_sum:
            return False
        for elem in matrix[i]:
            if elem not in nums_list:
                return False
            nums_list.remove(elem)


    return row_sum == sum([matrix[n - 1 - i][i] for i in range(n)])



n = int(input())
matrix = input_matrix(n)
print("YNEOS"[not check_magic(matrix, n)::2])
```

На вход программе подаются два натуральных числа n и m. Напишите программу, которая создает матрицу размером n×m, заполнив её "диагоналями" в соответствии с образцом: \\
**Sample Input 1:** \\
3 5 \\
**Sample Output 1:** \\
1  2  4  7  10 \\
3  5  8  11 13 \\
6  9  12 14 15 \\

---

```python
# put your python code here

def fill_matrix(n,m):
    matrix = [[1 for i in range(m)] for j in range(n)]
    counter = 0
    for i in range(n + m):
        for j in range(max(i - m + 1, 0),
                       max(i - m + 1, 0) + min(i + 1, n, m, n + m - i - 1)):
            counter += 1
            matrix[j][i - j] = counter
    return matrix

def print_matrix(matrix):
    for raw in matrix:
        print(*[str(elem).ljust(3) for elem in raw])

matrix_size = input().split()
n = int(matrix_size[0])
m = int(matrix_size[1])
matrix = fill_matrix(n,m)
print_matrix(matrix)

```

### Заполнение спиралью 😈😈

---
На вход программе подаются два натуральных числа n и m. Напишите программу, которая создает матрицу размером n×m, заполнив её "спиралью" в соответствии с образцом. \\
**Sample Input 1:** \\
4 5 \\
**Sample Output 1:** \\
1  2  3  4  5 \\
14 15 16 17 6 \\
13 20 19 18 7 \\
12 11 10 9  8 \\

---



```python
# put your python code here

def increase_to_right(matrix, step, counter, m):
    for j in range(step, m - 1 - step):
        if not matrix[step][j]:
            matrix[step][j] = counter
            counter += 1
        else:
            break

    return counter

def increase_to_down(matrix, step,counter, n, m):
    for i in range(step, n - 1 - step):
        if not matrix[i][m - 1 - step]:
            matrix[i][m - 1 -step] = counter
            counter += 1
        else:
            break
    return counter

def increase_to_left(matrix, step,counter, n, m):
    for j in range(m - 1 - step, step, -1):
        if not matrix[n - 1 - step][j]:
            matrix[n - 1 - step][j] = counter
            counter += 1
        else:
            break

    return counter

def increase_to_up(matrix, step,counter,n):
    for i in range(n - 1 - step, step, -1):
        if not matrix[i][step]:
            matrix[i][step] = counter
            counter += 1
        else:
            break

    return counter

def fill_matrix(n,m):
    matrix = [[0 for _ in range(m)] for _ in range(n)]
    counter = 1
    max_number = n * m + 1
    if m == n:
        max_number -= 1
        if n % 2:
            matrix[n // 2][m//2] = max_number
    step = 0
    while counter < max_number:
        counter = increase_to_right(matrix, step, counter, m)
        counter = increase_to_down(matrix, step, counter, n, m)
        counter = increase_to_left(matrix, step, counter, n, m)
        counter = increase_to_up(matrix, step, counter, n)
        step += 1


    return matrix



def print_matrix(matrix):
    for raw in matrix:
        print(*[str(elem).ljust(3) for elem in raw])

matrix_size = input().split()
n = int(matrix_size[0])
m = int(matrix_size[1])
matrix = fill_matrix(n,m)
print_matrix(matrix)
```

---


```python
def merge(values):
    res_dict = {}
    for dict in values:
        for key, value in dict.items():
            res_dict.setdefault(key,set()).add(value)

    return res_dict

print(merge([{'a': 1, 'b': 2}, {'b': 10, 'c': 100}, {'a': 1, 'b': 17, 'c': 50}, {'a': 5, 'd': 777}]))
```

* matrix() — возвращает матрицу 1×1, в которой единственное число равно нулю;
* matrix(n) — возвращает матрицу n×n, заполненную нулями;
* matrix(n, m) — возвращает матрицу из n строк и m столбцов, заполненную нулями;
* matrix(n, m, value) — возвращает матрицу из n строк и m столбцов, в которой каждый элемент равен числу value.

При создании функции пользуйтесь аргументами по умолчанию.

---

```python
def matrix(rows = 1, cells = None, value=0):
    if not cells:
        cells = rows

    return [[value for _ in range(cells)] for _ in range(rows)]

print(matrix())         # матрица 1 × 1 из 0
print(matrix(3))        # матрица 3 × 3 из 0
print(matrix(2, 5))     # матрица 2 × 5 из 0
print(matrix(3, 4, 9))  # матрица 3 × 4 из 9
```

На вход программе подается натуральное число n, а затем n строк английских слов в разных регистрах.

Напишите программу, которая выводит слова в начальном регистре (каждое на отдельной строке) в порядке возрастания их гематрии. Если гематрия слов совпадает, они выводятся в алфавитном (лексикографическом) порядке.

---

```python
# put your python code here
from functools import reduce
def calculate_gematry(word):
    return reduce(lambda x,char: x + ord(char.upper()) - ord('A'), word, 0)
print(*sorted([input() for _ in range(int(input()))], key=lambda word: (calculate_gematry(word),word)),sep='\n')
```

### Сортировка IP-адресов

---
IP-адрес – уникальный числовой идентификатор устройства в компьютерной сети, работающий по протоколу TCP/IP.

В 4-й версии IP-адрес представляет собой 32-битное число. Адрес записывается в виде четырёх десятичных чисел (октетов) со значением от 0 до 255, разделённых точками, например, 192.168.1.2.

Напишите программу, которая считывает IP-адреса и выводит их в порядке возрастания в соответствии с десятичным представлением.

---

```python
from functools import reduce
def calculate_ip_adress(ip_adress):
    return reduce(lambda acc, num: acc + int(num[0])*256**num[1], zip(ip_adress.split('.'), range(3,-1,-1)), 0)
print(*sorted([input() for _ in range(int(input()))],key=lambda ip_adress:calculate_ip_adress(ip_adress)), sep='\n')
```

### Сумма чисел в файле

---
Вам доступен текстовый файл nums.txt. В файле могут быть записаны целые неотрицательные числа и все, что угодно. Числом назовем последовательность одной и более цифр, идущих подряд (число всегда неотрицательно).

Напишите программу, которая вычисляет сумму всех чисел, записанных в файле.

---

```python
from functools import reduce
import string
with open("nums.txt", 'r', encoding='utf-8') as file:
    line = file.read().replace('\n',' ')
    nums_line = (reduce(lambda x,char: x + (' ',char)[char.isdigit()], line))
    print(reduce(lambda acc, num: acc + int(num), nums_line.split(), 0))
```

### CSV-файл

---
Вам доступен CSV-файл data.csv, содержащий информацию в csv формате. Напишите функцию read_csv для чтения данных из этого файла. Она должна возвращать список словарей, интерпретируя первую строку как имена ключей, а каждую последующую строку как значения этих ключей.

image.png

---

```python
def read_csv():
    data_info = []

    with open("data.csv") as data:
        data_keys = data.readline().strip('\n').split(',')
        data_info = []
        for info in data:
            info = info.strip('\n').split(',')
            data_info.append(dict(zip(data_keys, info)))

    return data_info
```

### Загадка от Жака Фреско 🌶️

---
Однажды Жака Фреско спросили:

"Если ты такой умный, почему не богатый?"

Жак не стал отвечать на столь провокационный вопрос, вместо этого он задал загадку спрашивающему:

"Были разноцветные козлы. Сколько?"

"Сколько чего?"

"Сколько из них составляет более 7% от общего количества козлов?"

Вам доступен текстовый файл goats.txt в первой строке которого написано слово COLOURS, далее идет список всех возможных цветов козлов. Затем идет строка со словом GOATS, и далее непосредственно перечисление козлов разных цветов. Перечень козлов включает только строки из первого списка.

Напишите программу создания файла answer.txt и вывода в него списка козлов, которые удовлетворяют условию загадки от Жака Фреско.

---

```python
# put your python code here
from decimal import Decimal
with open("goats.txt", 'r', encoding='utf-8') as goats_info, open("answer.txt", 'w', encoding='utf-8') as answer:
    lines = list(map(lambda info : info.strip('\n'), goats_info.readlines()))
    goats_count = len(lines[lines.index('GOATS') + 1::])
    goats_colors = dict.fromkeys(lines[1:lines.index('GOATS')])
    goats_colors = {goat : lines.count(goat) - 1 for goat in goats_colors}
    answer_goats = [goat for goat, count in goats_colors.items() if count > Decimal(0.07 * goats_count)]
    print(*sorted(answer_goats), file=answer, sep='\n')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Label', axis=1)
```

```python
y = df['Label']
```

```python
X.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
from sklearn.preprocessing import StandardScaler, Normalizer
```

```python
scaler = Normalizer()
```

```python
from sklearn.pipeline import Pipeline
```

```python
steps = [('scaler', scaler), ('knn', knn)]
```

```python
k_values = list(range(1,30))
```

```python
knn_param_grid = {'knn__n_neighbors':k_values}
```

```python
pipe = Pipeline(steps)
```

```python
pipe.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
grid_model = GridSearchCV(pipe, 
                          param_grid=knn_param_grid, 
                          cv=10,
                          scoring='accuracy',
                          verbose=3)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.get_params()
```

```python
grid_model.cv_results_['mean_test_score']
```

```python
plt.plot(list(range(1,30)), grid_model.cv_results_['mean_test_score'], marker='o')
plt.grid(True)
plt.xlabel('K')
plt.ylabel('Accuracy')
```

```python
y_pred = grid_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(grid_model, X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```
